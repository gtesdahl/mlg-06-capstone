# Data Acquisition and Exploration

### Imports

```python
import pandas as pd
from bs4 import BeautifulSoup
import re
import requests
import urllib.request
import imageio
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
from IPython.display import clear_output
from os import listdir
from os.path import isfile, join
import re
```

### Import goodreads reviews csv

```python
goodreads = pd.read_csv('books.csv', error_bad_lines=False)

goodreads.head()
```

### Identify the covers from the goodreads dataframe that still need to be downloaded

```python
jpg_list = [f for f in listdir('cover_images\\') if isfile(join('cover_images\\', f))]

downloaded_images = pd.Series(list(map(int,[i[:-4] for i in jpg_list])))
scrape_list = goodreads[~goodreads['bookID'].isin(downloaded_images)]['bookID']
```

```python
len(downloaded_images)
```

### Download the cover for each book in the goodreads dataframe

```python
for i, book_id in enumerate(scrape_list):
    clear_output(wait=True)
    
    url = "https://www.goodreads.com/book/show/" + str(book_id)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')

    try:
        bk = soup.find(id="imagecol").find(class_="bookCoverPrimary").find('img')['src']
        urllib.request.urlretrieve(bk, "cover_images\\"+str(book_id)+".jpg")
        
        print("Current progress: "+ str(round((i / len(scrape_list))*100,2)) + "%")
    
    except AttributeError:
        print("Current progress: " + str(round((i / len(scrape_list))*100,2)),"%")
```

### Create additional fields for the images

```python
#empty dataframe
pic_features = pd.DataFrame(columns=['Book ID', 'Pic Height', 'Pic Width', 'Pic Dimension', 'Pic Size', 'Avg R', 'Avg B', 'Avg G'])

#appends the features of each image
for image_name in downloaded_images:
    
    try:
        pic = imageio.imread('cover_images/' + str(image_name) + '.jpg')
    
    except ValueError:
        print(image_name)
        continue
    
    if pic.ndim == 3:
        df = pd.DataFrame({'bookID': [image_name], 'Pic Height': [pic.shape[0]], 'Pic Width': [pic.shape[1]], 
                           'Pic Dimension': [pic.ndim], 'Pic Size': [pic.size], 'Avg R':[pic[:, :, 0].mean()], 
                           'Avg B':[pic[:, :, 1].mean()], 'Avg G':[pic[:, :, 2].mean()]})
        pic_features = pic_features.append(df)

pic_features.sort_values('bookID', inplace=True)
pic_features.reset_index(inplace=True, drop=True)
pic_features.head()
```

### Merge the goodreads reviews with the book cover features

```python
cover_df = goodreads.merge(pic_features, left_on='bookID', right_on='bookID', how='inner')
cover_df.head()
```

```python
cover_df.shape
```

### Analyze the differences in image aspect ratios

```python
cover_df['Aspect Ratio 1'] = cover_df['Pic Height'] / cover_df['Pic Width']
cover_df['Aspect Ratio 2'] = cover_df['Pic Width'] / cover_df['Pic Height']
cover_df['Aspect Ratio Max'] = cover_df[['Aspect Ratio 1', 'Aspect Ratio 2']].max(axis=1)
```

# v2 - Collect new book data

```python
# location to store images
image_loc = 'cover_images\\'

# load dataframes
try:
    book_df = pd.read_csv('books_updated.csv')
except:
    book_df = pd.DataFrame(columns=['bookID', 'average_rating', 'ratings_count', 'summary'])
    
# try:
#     failed_ids = pd.read_csv('failed_ids.csv')
# except:
#     failed_ids = pd.DataFrame(columns=['bookID', 'kind'])

# set of potential ids
potential_book_ids = {i for i in range(1,5000)}

# book ids with images already gathered
jpg_list = [f for f in listdir(image_loc) if isfile(join(image_loc, f))]
book_ids_with_images = set(map(int,[i[:-4] for i in jpg_list]))
book_ids_with_data = set(book_df[~book_df['summary'].isnull()]['bookID'].values)
```

```python
# identify book ids to scrape
skip_failed_downloads = False
skip_downloaded_images = False
skip_downloaded_data = True

if skip_failed_downloads:
    failed_ids = set(failed_ids['bookID'].values)
    potential_book_ids = potential_book_ids.difference(failed_ids)
    
if skip_downloaded_images:
    potential_book_ids = potential_book_ids.difference(book_ids_with_images)
    
if skip_downloaded_data:
    potential_book_ids = potential_book_ids.difference(book_ids_with_data)

book_ids_with_images = list(book_ids_with_images)
scrape_list = list(potential_book_ids)
```

```python
[len(i) for i in [book_df, potential_book_ids, book_ids_with_images, scrape_list]]
```

```python
# wrap in for loop x5 that rests for 10 minutes between loops

# track successes & failures
successes_images = 0
failures_images = 0

successes_ratings = 0
failures_ratings = 0

successes_summaries = 0
failures_summaries = 0

format_pct = lambda x, y: round((x / y)*100,2)


# scrape goodreads
for book_id in scrape_list:
    clear_output(wait=True)
    
    # pull webpage
    url = "https://www.goodreads.com/book/show/" + str(book_id)
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Safari/537.36"}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')

    # only scrape image if not already downloaded
    if book_id not in book_ids_with_images:
        try:
            bk = soup.find(id="imagecol").find(class_="bookCoverPrimary").find('img')['src']
            urllib.request.urlretrieve(bk, image_loc + str(book_id) + ".jpg")
            successes_images += 1
            
        except:        
#             failed_ids = failed_ids.append({'bookID':book_id, 'kind':'image'}, ignore_index=True)
            failures_images += 1
        
    # only scrape ratings if blank in csv
    if book_id not in book_df['bookID'].values:
        try:
            average_rating = soup.find('span', itemprop="ratingValue").string.strip()
            average_rating = float(average_rating)

            ratings_count = soup.find('a', href="#other_reviews").find('meta', itemprop="ratingCount").string.strip()
            ratings_count = int(re.sub('[^0-9]','', ratings_count))

            new_row = {'bookID':book_id, 'average_rating':average_rating, 'ratings_count':ratings_count, 'summary':''}
            book_df = book_df.append(new_row, ignore_index=True)
            
            successes_ratings += 1

        except:        
#             failed_ids = failed_ids.append({'bookID':book_id, 'kind':'ratings'}, ignore_index=True)

            failures_ratings += 1
      
    # only scrape summaries if the bookID exists in csv
    if book_id in book_df['bookID'].values:
        try:
            summary = soup.find('div', id="descriptionContainer").find('div', id='description').find('span', style='display:none').text
            book_df[book_df['bookID'] == book_id]['summary'] = summary

            successes_summaries += 1

        except:        
#                 failed_ids = failed_ids.append({'bookID':book_id, 'kind':'summary'}, ignore_index=True)

            failures_summaries += 1

        
    count_ratings = max(successes_ratings + failures_ratings,1)
    count_images = max(successes_images + failures_images,1)
    count_summaries = max(successes_summaries + failures_summaries,1)
    count = max(count_ratings, count_images, count_summaries)
    total = len(scrape_list)
    
    print(f"Current progress:\t {count} of {total} bookIDs ({format_pct(count, total)}%)")
    print(f'Image Successes:\t {successes_images} ({format_pct(successes_images, count_images)}%)')
    print(f'Image Failures:\t\t {failures_images} ({format_pct(failures_images, count_images)}%)')
    print(f'Data Successes:\t\t {successes_ratings} ({format_pct(successes_ratings, count_ratings)}%)')
    print(f'Data Failures:\t\t {failures_ratings} ({format_pct(failures_ratings, count_ratings)}%)')
    print(f'Summary Successes:\t {successes_summaries} ({format_pct(successes_summaries, count_summaries)}%)')
    print(f'Summary Failures:\t {failures_summaries} ({format_pct(failures_summaries, count_summaries)}%)')

# export datafromes
book_df.to_csv('C:\\Users\\gtesdahl\\Documents\\GitHub\\mlg-06-capstone\\books_updated.csv', index=False)
# failed_ids.to_csv('C:\\Users\\gtesdahl\\Documents\\GitHub\\mlg-06-capstone\\failed_ids.csv', index=False)
```
